```python
#IMPORTAR neomdel
from neomodel import StructuredNode, StringProperty, RelationshipTo, RelationshipFrom, config, IntegerProperty, UniqueIdProperty, UniqueIdProperty

#URL CONECCION CON LA BASE DE DATOS DE NEO4J
config.DATABASE_URL = 'bolt://neo4j:Proyecto@localhost:11003'

#CREAR Object PAIS
class PAIS(StructuredNode):
    nombre = StringProperty(unique_index=True)
    datos = RelationshipFrom('UBICACION','CIUDAD')

#CREAR Object USUARIOS
class USUARIOS(StructuredNode):
    nombre = StringProperty(unique_index=True)
    datos = RelationshipTo('DATOS','DATOS')
    ubicacion = RelationshipTo('UBICACION','UBICACION')
    
#CREAR Object UBICACION 
class UBICACION(StructuredNode):
    location = StringProperty(unique_index=True)
    usuarios = RelationshipFrom('USUARIOS','UBICACION')
    pais = RelationshipTo('PAIS','CIUDAD') 

#CREAR Object DATOS 
class DATOS(StructuredNode):
    fecha = StringProperty(unique_index=True)
    url = StringProperty(unique_index=True)
    texto = StringProperty(unique_index=True)
    publicado_desde = StringProperty(unique_index=True)
    favorite_count = StringProperty(unique_index=True)
    retweet_count = StringProperty(unique_index=True)
    usuarios = RelationshipFrom('USUARIOS','DATOS')
```
